Применение критерия Шапиро-Уилка

```python
res = stats.shapiro(np.log10(data[u'население']))
print('p-value:', res[1])
```

```python
plt.hist(np.log10(data['население']), bins = 50)
```

Если убрать три выброса, то критерий шапира-уилка все равно отвергнет гипотезу нормальности, но в принципе, после того, как их убрать можно считать критерий стьюдента и корреляцию, так как отклонения от нормальности незначительны

```python
data = pd.read_csv('Albuquerque Home Prices.txt', sep = '\t')
data
```

```python
data.replace(-9999, np.nan, inplace = True)
data
```

```python
x = data[data['COR'] == 1]['PRICE']
y = data[data['COR'] == 0]['PRICE']
x.name = 'corner'
y.name = 'not corner'
print(x)
print(y)
```

```python
def two_hist(x, y):
    x.hist(alpha = 0.5, weights = [1. / len(x)] * len(x), color = 'red')
    y.hist(alpha = 0.5, weights = [1. / len(y)] * len(y), color = 'blue')
    plt.axvline(x.mean(), color = 'red', alpha = 0.8, linestyle = 'dashed')
    plt.axvline(y.mean(), color = 'blue', alpha = 0.8, linestyle = 'dashed')
    plt.legend([x.name, y.name])
```

```python
two_hist(x, y)
```

Существенные отличия от нормальности, идея сравнивать выборки по средним не подошла, значит будем сравнивать медианы с помощью критерия манна-уитни

```python
res = stats.mannwhitneyu(x, y)
print('p_value:', res[1])
```

```python
data = pd.read_csv('agedeath.dat.txt', sep = '\s+', header = None, names = ['group', 'age', 'index'])
data
```

```python
x = data[data['group'] == 'sovr']['age']
y = data[data['group'] == 'aris']['age']
x.name = 'sovr'
y.name = 'aris'
```

```python
two_hist(x, y)
```

```python
res = stats.fligner(x, y)
print('p-value:', res[1])
```

```python
res = stats.ttest_ind(x, y, equal_var = False)
print('p-value:', res[1])
```

Критерий Стьюдента для парных выборок

```python
data = pd.read_csv('interference.csv')
data
```

```python
x = data['DiffCol']
y = data['Black']
x.name = 'Diffcol'
y.name = 'Black'
```

```python
two_hist(x, y)
```

```python
res = stats.ttest_rel(x, y)
print('p-value:', res[1])
```

```python
con_table = pd.DataFrame([[28, 72], [20, 80]], index = ['city', 'country'], columns = ['for', 'againts'])
con_table
```

```python
res = stats.chi2_contingency(con_table)
print('p-value:', res[1])
```

```python
data = pd.read_csv('Albuquerque Home Prices.txt', sep = '\t')
data.replace(-9999, np.nan)
data
```

```python
plt.scatter(data['PRICE'], data['SQFT'])
```

```python
res = stats.pearsonr(data['PRICE'], data['SQFT'])
print('p-value:', res[1])
```
